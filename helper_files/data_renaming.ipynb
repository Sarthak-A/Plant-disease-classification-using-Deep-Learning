```python
import os
import shutil

def rename_images(root_dir):
    # Iterate through all subdirectories
    for class_dir in os.listdir(root_dir):
        class_path = os.path.join(root_dir, class_dir)
        if os.path.isdir(class_path):
            # Iterate through images in each subdirectory
            images = [f for f in os.listdir(class_path) if os.path.isfile(os.path.join(class_path, f))]
            for i, image in enumerate(images):
                # Rename each image
                _, ext = os.path.splitext(image)
                new_name = f"{class_dir}_{i}{ext}"
                os.rename(os.path.join(class_path, image), os.path.join(class_path, new_name))
            print(f"Renamed {len(images)} images in {class_dir}")

def copy_folders_with_renamed_images(source_dir, destination_dir):
    # Copy the folder to the destination directory
    shutil.copytree(source_dir, destination_dir)
    # Rename images in the copied folder
    rename_images(destination_dir)

# Example usage:
source_directory = r"C:\Users\ankit\PDMD"
destination_directory = r"C:\Users\ankit\PDMD 2"

copy_folders_with_renamed_images(source_directory, destination_directory)


print('Renaming complete')
```
